```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
price = pd.read_csv('agrodata.csv')
price.head()
```

```python
price['Date'] = price['Date'].astype(str)
```

```python
price['Date'] = pd.to_datetime(price['Date'], format = "%d/%m/%Y")
```

```python
price.info()
```

```python
weather = pd.read_csv('agroweather_data.csv')
weather.head()
```

```python
weather.drop(['Unnamed: 0', 'bateria', 'fechaUtlMod', 'et0'], axis = 1, inplace =True)
weather.head()
```

```python
weather['fecha'] = pd.to_datetime(weather['fecha'])
weather.info()
```

```python
weather.describe()
```

```python
final = price.merge(weather, how = 'inner', left_on = 'Date', right_on = 'fecha')
final.head()
```

```python
final['Date'] = final['Date'].astype(str)
```

```python
final['Dia'] = final['Date'].str.split('-').str[2]
final['Mes'] = final['Date'].str.split('-').str[1]
final['Año'] = final['Date'].str.split('-').str[0]
```

```python
final.drop(['Date', 'fecha'], axis = 1, inplace = True)
final.head()
```

```python
def custom_round(time_str):
    # Asegurarse de que time_str es una cadena
    time_str = str(time_str)
    
    # Verificar si time_str contiene ':'
    if ':' in time_str:
        # Dividir en horas y minutos
        hour, minute = map(int, time_str.split(':'))
    else:
        # Si no contiene ':', asumir que es solo una hora
        hour = int(time_str)
        minute = 0  # No hay minutos especificados
    
    # Redondear la hora basándose en el valor de los minutos
    return hour + (minute >= 30)
```

```python
final.isnull().sum()
```

```python
final['horMinTempMax'] = final['horMinTempMax'].astype(str).apply(custom_round)
final['horMinTempMin'] = final['horMinTempMin'].astype(str).apply(custom_round)
final['horMinHumMin'] = final['horMinHumMin'].astype(str).apply(custom_round)
final['horMinVelMax'] = final['horMinVelMax'].astype(str).apply(custom_round)
final['horMinVelMax'] = final['horMinVelMax'].astype(str).apply(custom_round)
final['horMinHumMax'] = final['horMinHumMax'].astype(str).apply(custom_round)
```

```python
final.drop(['Dia', 'Mes', 'Año'], axis = 1, inplace =True)
final.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X = final.drop('Price', axis = 1)
y = final['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=123)
```

```python
X_train.shape, y_train.shape
```

```python
scaler.fit_transform(X_train)
scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, accuracy_score, mean_absolute_percentage_error, mean_absolute_error
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor
```

```python
rf = RandomForestRegressor(n_estimators = 10000, min_samples_leaf=0.2)
```

```python
dt = DecisionTreeRegressor(random_state=123)
```

```python
rf.fit(X_train, y_train)
predictions = rf.predict(X_test)
```

```python
result = mean_absolute_error(predictions, y_test)
result
```

```python
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test)
```

```python
dt_result = mean_absolute_error(dt_pred, y_test)
print(f"El error que obtenemos con DecisionTree es de {round((dt_result/100), 2)}€")
```
